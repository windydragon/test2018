```python
import numpy as np

def deviation(x,y,m,n):
    deviation = max(x/m-y/n, y/n-x/m)
    return deviation

m = 11.
n = 7.

deviation_list = []
for i in range(int(m)+1):
    for j in range(int(n)+1):
        deviation_list.append(deviation(i, j, m, n))

mean_deviation = np.mean(deviation_list)
print mean_deviation
std_deviation = np.std(deviation_list)
print std_deviation
```
